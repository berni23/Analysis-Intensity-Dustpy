## Reading in from default format

Imports for this part:

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.style.use(['seaborn',{'figure.dpi':150}])
```

Loading the data

```python
data = np.load('default_opacities.npz')

a = data['a']
lam = data['lam']
k_abs = data['k_abs']
k_sca = data['k_sca']
g = data['g']
```

```python
print(len(lam))
print (k_abs[0])

```

As an example: plot the opacities of the smallest grains as function of wavelength:

```python
f, ax = plt.subplots()
ax.loglog(lam, k_abs[:, 50], label='absorption')
ax.loglog(lam, k_sca[:, 50], label='scattering')
ax.set_title('particle size: {a[0]}')
ax.set_ylabel('opacity [cm$^2$ g$^{-1}$]')
ax.set_xlabel('$\lambda$ [cm]')
ax.legend();
```
