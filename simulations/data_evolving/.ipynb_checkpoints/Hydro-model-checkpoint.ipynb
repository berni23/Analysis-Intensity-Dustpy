# Hydro model data



Outputs are made every 20 orbits for the outermost planet at 165au. This correspond to about 28,000 yrs.

- `domain_y.dat` contains the radial grid positions in centimeters.
- `dens000000X.txt` contains the gas surface density after $X \times 20$ orbits for earch radial grid point.

```python
#Read the data
#13.10.18/17.10.18

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
a = np.zeros((51,1024),dtype=float)
R = np.loadtxt("domain_y.dat") # Radial grid positions



for i in range(0,51): #Array with a[i] = i*20 orbits
    
    if i<10:

        a[i] = np.loadtxt("dens000000"+ str(i) +  ".txt")

    else:
    
        a[i] = np.loadtxt("dens00000"+ str(i)  + ".txt")
           

```

```python



plt.figure(figsize=(12,6))

plt.plot(R,a[0],'blue',label = '0')
plt.plot(R,a[1],'red',label = '20')
plt.plot(R,a[2],'green',label = '40')
plt.plot(R,a[3],'orange',label = '60')
plt.legend()
plt.ylabel('$\Sigma_g$')
plt.xlabel('Radial position (cm)')
plt.xlim(left=None, right=None)

plt.show()
```
