```python
import pdb
import numpy as np
import glob
from matplotlib import pyplot as plt
import pandas as pd
import os
import h5py
import dustpy
from dustpy import plotting
from functionsLast import dens_to_int

names_prime = np.asarray([str(f) for f in glob.glob('9_1_1*')])
for i in names_prime: 
      if str(i)=='videos' or str(i)=='__pycache__':    
            
            continue
       
      elif os.path.isdir(str(i)) ==True: 
    
        print(str(i)) 
        
        Ise = dens_to_int(str(i),50,opacities = 'default_opacities-1.npz')
        Int = Ise[3] 
        R   = Ise[0] 
        np.savetxt(str(i)+'/I_R_op1',(R,Int),delimiter=',', header="I,R",fmt='%1.8f')
        Ise = dens_to_int(str(i),50)
        Int = Ise[3] 
        R   = Ise[0]   
        np.savetxt(str(i)+'/I_R',(R,Int),delimiter=',', header="I,R",fmt='%1.8f')

        with h5py.File(str(i) + '/data0050.hdf5') as fC: 
             rC = fC['grid/r'][()] 
             mC = fC['grid/m'][()] 
             sig_dC = fC['dust/Sigma'][()] 
             csC = fC['gas/cs'][()] 
             alphaC = fC['gas/alpha'][()] 
             VfragC = fC['dust/vFrag'][()] 
             SigmaC = fC["gas/Sigma"][()] 
             omegaC = fC["grid/OmegaK"][()] 
             St = fC["dust/St"][()] 
             d2g = fC["dust/dust2gasRatio"][()] 
             rIntC = fC["grid/rInt"][()] 
             rhoC = fC['dust/rhoBulk'][()][0, 0] 
             agrain = fC['dust/a'][()][0] 
             Vk = omegaC * rC 
        
        FragBarrier = VfragC*VfragC / csC/csC/alphaC /3 
        FragBarrier *= 2 * SigmaC / np.pi /rhoC 
        p = SigmaC * omegaC * csC 

        from scipy.interpolate import interp1d 
        _f = interp1d(np.log10(rC), np.log10(p), fill_value='extrapolate') 
        pInt = 10.**_f(np.log10(rIntC)) 
        Diff_pint = np.diff(pInt) 
        Diff_rint = np.diff(rIntC) 
        gammaC = np.abs(rC / p * Diff_pint / Diff_rint) 
        DriftBarrier = 2 * d2g * SigmaC * Vk*Vk /csC/csC/gammaC/ np.pi /rhoC 
  
        np.savetxt(str(i)+'/rC.txt',rC,delimiter = ',')
        np.savetxt(str(i)+'/agrain.txt',agrain,delimiter = ',')
        np.savetxt(str(i)+'/sigma.txt',np.log10(sig_dC).T,delimiter = ',')
        np.savetxt(str(i)+'/Frag.txt',FragBarrier,delimiter = ',')
        np.savetxt(str(i)+'/Drift.txt',DriftBarrier,delimiter = ',')
        plotting.movie(str(i))
```

```python
"""Best fit HD profile"""

from astropy import constants as cc
from functionsLast import dens_to_int,c

HD = np.genfromtxt('HD163296.profile.txt')
lam_obs= 0.13
#r_HD = np.arctan(HD[:,0] * c.AU / d_n) * 60**2 * (180 / np.pi)
r_HD_au =  HD[:, 0]
I_HD = HD[:, 4] * 2 * lam_obs**-2 * cc.k_B.cgs.value # this should be in cgs
sigma_HD = HD[:,5]*2 * lam_obs**-2 * cc.k_B.cgs.value
theta_HD = np.arctan(HD[:, 0]*c.AU /101.5/c.pc) * 60**2 * (180 / np.pi)

names_prime = np.asarray([str(f) for f in glob.glob('*')])
names =[]
Min=[]
time=[]

Ise = dens_to_int('17_1_5',50)
theta_sim = Ise[0]

I_interp = np.interp(theta_sim[:136],theta_HD[:202],I_HD[:202])
sigma_interp =np.interp(theta_sim[:136],theta_HD[:202],theta_HD[:202])

def Chi2(X):
    
    return np.sum(((I_interp-X)*sigma_interp**-1)**2)
    
for i in names_prime:

    if str(i)=='videos' or str(i)=='__pycache__':  
        
        print('no')
        continue  
        
    if os.path.isdir(str(i)) ==True:
     
        names.append(str(i))

%time
for i in names:

    test =[]
    for j in range(0,51):
          
       Ise = dens_to_int(str(i),j)
       I_sim = Ise[2]

       chisquare = Chi2(I_sim[:136])
       test.append(chisquare)
     
    Min.append(np.min(test))
    time.append(test.index(Min[-1]))
    
    
best = np.min(Min)
print(names[Min.index(best)])
print(time[Min.index(best)])

I =dens_to_int(str(names[Min.index(best)]),time[Min.index(best)])

plt.semilogy(I[0],I[3])
plt.semilogy(theta_HD,I_HD)

%time  
#return theta_obs, I_isella, I_sim, I_sim_conv, r_HD, I_HD, HD_conv,r_sim,sigma_HD
```
